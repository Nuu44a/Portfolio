```python
'''
    2. Приветствие
'''
# Первый вариант
name = input('Как твое имя, бледнолицый? ')
s = 'Hello, ' + name + '!' + ' ' * 10
print(*[(s[i:] + s[:i]) * 3 for i in range(len(s), 0, -1)], sep='\n')


# Второй вариант
name = 'Monty Python'
s = 'Hello, ' + name + '!'
s = [i + ' ' for i in s]
s.append(' ' * 10)

for _ in range(len(s)):
    print(*s)
    rem_last = s.pop(0)
    s.append(rem_last)
```
